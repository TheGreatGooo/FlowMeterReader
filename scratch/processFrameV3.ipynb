```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

def show_wait_destroy(winname, img):
    cv.imshow(winname, img)
    cv.moveWindow(winname, 500, 0)
    cv.waitKey(0)
    cv.destroyWindow(winname)
    
```

```python
img2 = cv.imread('calibration/undistorted/ch0-1765830911.jpg')
```

```python
img1 = cv.imread('calibration/undistorted/ch0-1765835383.jpg')
```

```python
gray1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
gray2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
```

```python
diff = cv.absdiff(gray1, gray2)
thresh = cv.threshold(diff, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)[1]
cnts = cv.findContours(thresh.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
for c in cnts:
    x, y, w, h = cv.boundingRect(c)
    cv.rectangle(img1, (x, y), (x + w, y + h), (0, 0, 255), 2)
    cv.rectangle(img2, (x, y), (x + w, y + h), (0, 0, 255), 2)
cv.imshow("Image 1", img1)
cv.imshow("Image 2", img2)
cv.waitKey(0)
```
